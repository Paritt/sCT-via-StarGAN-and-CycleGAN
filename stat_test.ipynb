```python
import numpy as np
from scipy import stats
import pandas as pd
```

# Compare to CBCT

```python
model = 'CycleMRI'
df = pd.read_csv('case_report_nl.csv')
if model == 'CycleMRI' or model == 'StarMRI':
    name_to_drop = 'Case_22'
    df = df.query("Case != @name_to_drop")
    name_to_drop_2 = 'Case_60'
    df = df.query("Case != @name_to_drop_2")

with open(f'result/pp_abs_{model}_stat_result.txt', "w") as file:
    for i in range(2,28):
        metric = df.columns[i]
        CCB = np.abs(np.array(df[df['model'] == model][metric]))
        CBCT = np.abs(np.array(df[df['model'] == 'CBCT'][metric]))
        diff = CCB - CBCT
        shapiro_stat, shapiro_p = stats.shapiro(diff)
        if shapiro_p < 0.05:
            stat, p = stats.wilcoxon(CCB, CBCT)
            if p < 0.05:
                print(f'{metric} [Wilcoxon test] \nSignificant \n{round(p,4)}', file=file)
            else:
                print(f'{metric} [Wilcoxon test] \nNOT Significant \n{round(p,4)}', file=file)
        else:
            t_statistic, p = stats.ttest_rel(CCB , CBCT)
            if p < 0.05:
                print(f'{metric} [Pair T-test] \nSignificant \n{round(p,4)}', file=file)
            else:
                print(f'{metric} [Pair T-test] \nNOT Significant \n{round(p,4)}', file=file)
        print('\n', file=file)

with open(f'result/pp_abs_{model}_stat_result.txt', "r") as file:  # Open in read mode ("r")
    contents = file.read()  # Read entire content
    print(contents)
```

# Cycle vs Star

```python
model_ref = 'CycleCBCT'
model = 'StarCBCT'
df = pd.read_csv('case_report_nl.csv')
if model == 'CycleMRI' or model == 'StarMRI':
    name_to_drop = 'Case_22'
    df = df.query("Case != @name_to_drop")

with open(f'CycleVsStar/pp_abs_{model_ref}_vs_{model}_stat_result.txt', "w") as file:
    for i in range(2,28):
        metric = df.columns[i]
        eval = np.abs(np.array(df[df['model'] == model][metric]))
        ref = np.abs(np.array(df[df['model'] == model_ref][metric]))
        diff = eval - ref
        shapiro_stat, shapiro_p = stats.shapiro(diff)
        if shapiro_p < 0.05:
            stat, p = stats.wilcoxon(eval, ref)
            if p < 0.05:
                print(f'{metric} [Wilcoxon test] \nSignificant \n{round(p,4)}', file=file)
            else:
                print(f'{metric} [Wilcoxon test] \nNOT Significant \n{round(p,4)}', file=file)
        else:
            t_statistic, p = stats.ttest_rel(eval , ref)
            if p < 0.05:
                print(f'{metric} [Pair T-test] \nSignificant \n{round(p,4)}', file=file)
            else:
                print(f'{metric} [Pair T-test] \nNOT Significant \n{round(p,4)}', file=file)
        print('\n', file=file)

with open(f'CycleVsStar/pp_abs_{model_ref}_vs_{model}_stat_result.txt', "r") as file:  # Open in read mode ("r")
    contents = file.read()  # Read entire content
    print(contents)
```
